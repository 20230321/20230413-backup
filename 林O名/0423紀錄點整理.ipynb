```python
#NCC
def NormaliseCrossCorrelation(a,v):
    M = len(a)
    N = len(v)

    if type(a) is list: 
        a = np.array(a)
    if type(v) is list: 
        v = np.array(v)
    
    v_mu = np.mean(v)

    ncc_series = []
    for i in range(M-N+1):
        _a = a[i:i+N]
        a_sum = _a - np.mean(_a)
        v_sum = v - v_mu
        nr = 1/N * np.sum(a_sum*v_sum) / np.sqrt(np.var(_a)*np.var(v))
        ncc_series.append(nr)

    return ncc_series

def NormaliseCrossConvolution(a,v):
    M = len(a)
    N = len(v)

    if type(a) is list: 
        a = np.array(a)
    if type(v) is list: 
        v = np.array(v)
    
    v_mean = np.mean(v)
    v_normalized = v - v_mean
    a_normalized = a - np.mean(a)
    
    corr = np.convolve(a_normalized, v_normalized[::-1], mode='full')
    corr /= np.sqrt(np.sum(np.square(a_normalized)) * np.sum(np.square(v_normalized)))
    return corr

#A>B pname:圖片名稱備註 All:完整訊號 pain:疼痛7步訊號 nopain:不痛7步訊號 flag:判斷左右腳
def Rnopain_run_ncc(allname,all,nopain,find):



    #==================================不痛===========================================
        RX_nopain_plot=NormaliseCrossCorrelation(all,nopain)
        r_RX_nopain_plot=NormaliseCrossConvolution(all,nopain)
        r_RX_nopain_plot=r_RX_nopain_plot[:len(RX_nopain_plot)]
        turn_plot_start,turn_plot_end,one_plot_start,one_plot_end,two_plot_start,two_plot_end=forward_record_plot(len(all),len(nopain))
        # print("\n不痛最大")
        # print(max(RX_nopain_plot))
        # print("不痛平均")
        # print(np.average(RX_nopain_plot)) 
        # print("=====================================================================")
        
        # 創建一個figure和兩個subplot
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.6, 6))
        plt.subplots_adjust(hspace=0.3)


        # 第一個subplot顯示第一個圖表
        ax1.set_title("("+str(find)+"步"+")不痛Pattern 對上"+allname+"的相關性")
        ax1.plot(RX_nopain_plot)
        
        # for i in range(len (turn_plot_start)):
        #     if(i==0):
        #         ax1.axvline(x=turn_plot_start[i],c='green',label='turn')
        #         ax1.axvline(x=turn_plot_end[i],c='green')
        #     else:
        #         ax1.axvline(x=turn_plot_start[i],c='green')
        #         ax1.axvline(x=turn_plot_end[i],c='green')          


        for i in range(len (one_plot_end)):
            if(i==0):
                ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
                ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
            else:
                ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--")
                ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")


        for i in range(len (two_plot_start)):
            if(i==0):
                ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
                ax1.axvline(x=two_plot_end[i],c='r', linestyle="--") 
            else:
                ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
                ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")  
        
        ax1.legend(loc='upper right', shadow=True) 

        # 第二個subplot顯示第二個圖表
        ax2.set_title("("+str(find)+"步"+")"+"Convolution對上"+allname)
        #ax2.set_xlim([0,9600])
        ax2.plot(r_RX_nopain_plot)
        # for i in range(len (turn_plot_start)):
        #     if(i==0):
        #         ax2.axvline(x=turn_plot_start[i],c='green',label='turn')
        #         ax2.axvline(x=turn_plot_end[i],c='green')
        #     else:
        #         ax2.axvline(x=turn_plot_start[i],c='green')
        #         ax2.axvline(x=turn_plot_end[i],c='green')          


        for i in range(len (one_plot_end)):
            if(i==0):
                ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
                ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
            else:
                ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
                ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")


        for i in range(len (two_plot_start)):
            if(i==0):
                ax2.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
                ax2.axvline(x=two_plot_end[i],c='r', linestyle="--") 
            else:
                ax2.axvline(x=two_plot_start[i],c='r', linestyle="--")
                ax2.axvline(x=two_plot_end[i],c='r', linestyle="--")  
        
        ax2.legend(loc='upper right', shadow=True) 

        #=============================================================================================
        # 儲存並顯示圖表
        folder_path = "C:\\Users\\al574\\Desktop\\輩分\\林O名\\0423test\\術前\\左右腳不痛步態\\右腳"
        # 创建指定路径下的文件夹（如果它不存在的话）
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)



        # 保存图像到指定的文件夹中
        filename = os.path.join(folder_path, "("+str(find)+"步"+")不痛Pattern"+allname)
        #print(f"图片已保存为{filename}")
        plt.savefig(filename, facecolor='white', bbox_inches='tight', pad_inches=0.1)
        #plt.savefig("("+str(find)+"步"+")不痛Pattern"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
        plt.show()


            # 提示用户输入要保存的文件夹路径
        

        

#A>B pname:圖片名稱備註 All:完整訊號 pain:疼痛7步訊號 nopain:不痛7步訊號 flag:判斷左右腳
def Lnopain_run_ncc(allname,all,nopain,find):



    #==================================不痛===========================================
        RX_nopain_plot=NormaliseCrossCorrelation(all,nopain)
        r_RX_nopain_plot=NormaliseCrossConvolution(all,nopain)
        r_RX_nopain_plot=r_RX_nopain_plot[:len(RX_nopain_plot)]
        turn_plot_start,turn_plot_end,one_plot_start,one_plot_end,two_plot_start,two_plot_end=forward_record_plot(len(all),len(nopain))
        # print("\n不痛最大")
        # print(max(RX_nopain_plot))
        # print("不痛平均")
        # print(np.average(RX_nopain_plot)) 
        # print("=====================================================================")
        
        # 創建一個figure和兩個subplot
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.6, 6))
        plt.subplots_adjust(hspace=0.3)


        # 第一個subplot顯示第一個圖表
        ax1.set_title("("+str(find)+"步"+")不痛Pattern 對上"+allname+"的相關性")
        ax1.plot(RX_nopain_plot)
        
        # for i in range(len (turn_plot_start)):
        #     if(i==0):
        #         ax1.axvline(x=turn_plot_start[i],c='green',label='turn')
        #         ax1.axvline(x=turn_plot_end[i],c='green')
        #     else:
        #         ax1.axvline(x=turn_plot_start[i],c='green')
        #         ax1.axvline(x=turn_plot_end[i],c='green')          


        for i in range(len (one_plot_end)):
            if(i==0):
                ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
                ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
            else:
                ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--")
                ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")


        for i in range(len (two_plot_start)):
            if(i==0):
                ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
                ax1.axvline(x=two_plot_end[i],c='r', linestyle="--") 
            else:
                ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
                ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")  
        
        ax1.legend(loc='upper right', shadow=True) 

        # 第二個subplot顯示第二個圖表
        ax2.set_title("("+str(find)+"步"+")"+"Convolution對上"+allname)
        #ax2.set_xlim([0,9600])
        ax2.plot(r_RX_nopain_plot)
        # for i in range(len (turn_plot_start)):
        #     if(i==0):
        #         ax2.axvline(x=turn_plot_start[i],c='green',label='turn')
        #         ax2.axvline(x=turn_plot_end[i],c='green')
        #     else:
        #         ax2.axvline(x=turn_plot_start[i],c='green')
        #         ax2.axvline(x=turn_plot_end[i],c='green')          


        for i in range(len (one_plot_end)):
            if(i==0):
                ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
                ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
            else:
                ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
                ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")


        for i in range(len (two_plot_start)):
            if(i==0):
                ax2.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
                ax2.axvline(x=two_plot_end[i],c='r', linestyle="--") 
            else:
                ax2.axvline(x=two_plot_start[i],c='r', linestyle="--")
                ax2.axvline(x=two_plot_end[i],c='r', linestyle="--")  
        
        ax2.legend(loc='upper right', shadow=True) 

        #=============================================================================================
        # 儲存並顯示圖表
        folder_path = "C:\\Users\\al574\\Desktop\\輩分\\林O名\\0423test\\術前\\左右腳不痛步態\\左腳"
        # 创建指定路径下的文件夹（如果它不存在的话）
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)



        # 保存图像到指定的文件夹中
        filename = os.path.join(folder_path, "("+str(find)+"步"+")不痛Pattern"+allname)
        #print(f"图片已保存为{filename}")
        plt.savefig(filename, facecolor='white', bbox_inches='tight', pad_inches=0.1)
        #plt.savefig("("+str(find)+"步"+")不痛Pattern"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
        plt.show()


            # 提示用户输入要保存的文件夹路径
        

       



```

```python
R_flag=[]
#NCC
def NormaliseCrossCorrelation(a,v):
    M = len(a)
    N = len(v)

    if type(a) is list: 
        a = np.array(a)
    if type(v) is list: 
        v = np.array(v)
    
    v_mu = np.mean(v)

    ncc_series = []
    for i in range(M-N+1):
        _a = a[i:i+N]
        a_sum = _a - np.mean(_a)
        v_sum = v - v_mu
        
        nr = 1/N * np.sum(a_sum*v_sum) / np.sqrt(np.var(_a)*np.var(v))
        ncc_series.append(nr)

    return ncc_series


def NormaliseCrossConvolution(a,v):
    M = len(a)
    N = len(v)

    if type(a) is list: 
        a = np.array(a)
    if type(v) is list: 
        v = np.array(v)
    
    v_mean = np.mean(v)
    a_normalized = a - np.mean(a)
    v_normalized = v - v_mean
    #因為老師說要不一樣所以 mode=full 實際上是一樣的 因為是v翻轉不是實際波型所以相關度下降是正常的
    corr = np.convolve(a_normalized, v_normalized[::-1], mode='void')
    corr /= np.sqrt(np.sum(np.square(a_normalized)) * np.sum(np.square(v_normalized)))
    return corr



#A>B pname:圖片名稱備註 All:完整訊號 pain:疼痛7步訊號 nopain:不痛7步訊號 find:畫圖title用,listCorrelation:計算重複點,listConvolution:計算重複點
#flag:在哪種特徵會出現的位置,list_findpeaks_point:記錄用特徵位置
def one_run_ncc(allname,pname,all,pain,nopain,find,listCorrelation,listConvolution,flag,list_findpeaks_point):

    #先記錄findpeak之前的位置
    listoddpoint=[]
    listoddpoint=listCorrelation[:]
    # if(listCorrelation==listoddpoint):
    #     print("可以")
    RX_pain_plot=NormaliseCrossCorrelation(all,pain)
    r_RX_pain_plot=NormaliseCrossConvolution(all,pain)
    #one_plot_start,one_plot_end=one_plot(len(all),len(pain))
    turn_plot_start,turn_plot_end,one_plot_start,one_plot_end,two_plot_start,two_plot_end=forward_record_plot(len(all),len(pain))
    #r_turn_plot_start,r_turn_plot_end,r_one_plot_start,r_one_plot_end,r_two_plot_start,r_two_plot_end=reverse_record_plot(len(all),len(pain))    


    #print(oneplot)
    #print(len(RX_pain_plot))
    #RX_pain_plot=NormaliseCrossCorrelation(listRX1,RX_pain)
    # for i in range(len(RX_pain_plot)):
    #     if(RX_pain_plot[i]<0):
    #         RX_pain_plot[i]=RX_pain_plot[i]*-1   

    #print(RX_pain_plot)
    


    # print("\n疼痛最大")
    # print(max(RX_pain_plot))
    # print("疼痛平均")
    # print(np.average(RX_pain_plot))

    
    #全部
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.6, 6))
    plt.subplots_adjust(hspace=0.3)

    # 第一個subplot顯示第一個圖表
    ax1.set_title(pname+"("+str(find)+"步"+")疼痛Pattern 對上"+allname+"的相關性")
    #ax1.set_xlim([5750, 7000])
    #ax1.plot(RX_nopain_plot)
    RX_pain_plot=np.array(RX_pain_plot)
    RX_pain_plot_peaks, _= find_peaks(RX_pain_plot,height=0.5,distance=80)

    ax1.plot(RX_pain_plot)
    ax1.plot(RX_pain_plot_peaks, RX_pain_plot[RX_pain_plot_peaks],"*")

    for i in range(len (one_plot_end)):
        if(i==0):
            ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
            ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
        else:
            ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--")
            ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    for i in range(len (two_plot_end)):
        if(i==0):
            ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
            ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")            
        else:
            ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
            ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")
    
    ax1.legend(loc='upper left', shadow=True) 

    # 第二個subplot顯示第二個圖表
    ax2.set_title(pname+"("+str(find)+"步"+")"+"Convolution對上"+allname)
    #ax2.set_xlim([5750, 7000])
    

    r_RX_pain_plot=np.array(r_RX_pain_plot)
    r_RX_pain_plot_peaks, _= find_peaks(r_RX_pain_plot,height=0.5,distance=80)

    ax2.plot(r_RX_pain_plot)
    ax2.plot(r_RX_pain_plot_peaks, r_RX_pain_plot[r_RX_pain_plot_peaks],"*")



    for i in range(len (one_plot_end)):
        if(i==0):
            ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
            ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
        else:
            ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
            ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    for i in range(len (two_plot_end)):
        if(i==0):
            ax2.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
            ax2.axvline(x=two_plot_end[i],c='r', linestyle="--")            
        else:
            ax2.axvline(x=two_plot_start[i],c='r', linestyle="--")
            ax2.axvline(x=two_plot_end[i],c='r', linestyle="--")

    ax2.legend(loc='upper left', shadow=True) 
    # 儲存並顯示圖表

    # folder_name = "example_folder"

    # if not os.path.exists(folder_name):
    #     os.mkdir(folder_name)

    # if pname == '右腳X特徵':
    #     if  allname=="右腳X":
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/右腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)

    #     if  allname == '左腳X':
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/左腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)

    # if pname == '左腳X特徵':
    #     if  allname=="右腳X":
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/右腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)

    #     if  allname == '左腳X':
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/左腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)
    # 儲存並顯示圖表
    folder_path = "C:\\Users\\al574\\Desktop\\輩分\\林O名\\0423test\\術前\\左右腳一級疼痛步態\\"+str(allname)
    # 创建指定路径下的文件夹（如果它不存在的话）
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)



    # 保存图像到指定的文件夹中
    filename = os.path.join(folder_path, "("+str(find)+"步"+")一級Pattern"+allname)
    #print(f"图片已保存为{filename}")
    plt.savefig(filename, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #plt.savefig("("+str(find)+"步"+")不痛Pattern"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    plt.show()




    
#==============================================================================================
    # #縮放
    # fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 6))
    # plt.subplots_adjust(hspace=0.3)


    # # 第一個subplot顯示第一個圖表
    # ax1.set_title("("+str(find)+"步"+")疼痛Pattern 對上"+allname+"的相關性")
    # ax1.set_xlim([5750, 7000])
    # #ax1.plot(RX_nopain_plot)
    # RX_pain_plot=np.array(RX_pain_plot)
    # RX_pain_plot_peaks, _= find_peaks(RX_pain_plot,height=0.5,distance=80)

    # ax1.plot(RX_pain_plot[:9600])
    # ax1.plot(RX_pain_plot_peaks, RX_pain_plot[RX_pain_plot_peaks],"*")
    # for i in range(len (one_plot_end)):
    #     if(i==0):
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
    #     else:
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    # for i in range(len (two_plot_end)):
    #     if(i==0):
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")            
    #     else:
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")
    # ax1.legend(loc='upper right', shadow=True) 
    
  

    # # 第二個subplot顯示第二個圖表
    # ax2.set_title("("+str(find)+"步"+")"+"Convolution對上"+allname)
    # ax2.set_xlim([5750, 7000])
    

    # r_RX_pain_plot=np.array(r_RX_pain_plot)
    # r_RX_pain_plot_peaks, _= find_peaks(r_RX_pain_plot,height=0.5,distance=80)

    # ax2.plot(r_RX_pain_plot[:9600])
    # ax2.plot(r_RX_pain_plot_peaks, r_RX_pain_plot[r_RX_pain_plot_peaks],"*")

    # for i in range(len (one_plot_end)):
    #     if(i==0):
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
    #     else:
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    # for i in range(len (two_plot_end)):
    #     if(i==0):
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")            
    #     else:
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")
            
    # ax2.legend(loc='upper right', shadow=True) 
    # # 儲存並顯示圖表
    # #plt.savefig("("+str(find)+"步"+")簡化"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    # plt.show()
#==============================================================================================

    #計算重複點
    for i in range(len(RX_pain_plot_peaks)):
        listCorrelation[RX_pain_plot_peaks[i]]=listCorrelation[RX_pain_plot_peaks[i]]+1


    for i in range(len(r_RX_pain_plot_peaks)):
        listConvolution[r_RX_pain_plot_peaks[i]]=listConvolution[r_RX_pain_plot_peaks[i]]+1 
    

    for i in range(len(listCorrelation)):
        if(listCorrelation[i]!=listoddpoint[i]):
            # print(listCorrelation[i])
            # print(listoddpoint[i])
            list_findpeaks_point[i]=flag
            # print(i)
            # print("flag")
            # print(flag)

    # for i in range(len(RX_pain_plot_peaks)):
    #     if(listCorrelation[i]>1) :
    #         print(listCorrelation[i])
    #         print(listoddpoint[i]) 



    




    #==================================不痛===========================================
    # if(find==1): 
    #     RX_nopain_plot=NormaliseCrossCorrelation(all,nopain)
    #     #RX_nopain_plot=NormaliseCrossCorrelation(listRX1,RX_nopain)
    #     for i in range(len(RX_nopain_plot)):
    #         if(RX_nopain_plot[i]<0):
    #             RX_nopain_plot[i]=RX_nopain_plot[i]*-1
    #     # print("\n不痛最大")
    #     # print(max(RX_nopain_plot))
    #     # print("不痛平均")
    #     # print(np.average(RX_nopain_plot)) 
    #     # print("=====================================================================")




    #     print(pname)
    #     plt.title("不痛Pattern 對上"+allname+"的相關性")
    #     plt.plot(RX_nopain_plot)
    #     #plt.axvline(x=flag,c='r')
    #     #plt.plot(Rc_turn_pic)
    #    # plt.plot(Rc_one_pic)
        
    #     #plt.savefig(pname+'不疼痛.png', dpi=750,facecolor='white')
    #     #plt.axvline(x=listflag_one[0],c='r')
    #     #plt.axvline(x=63,c='r')
    #     plt.show()


#A>B pname:圖片名稱備註 All:完整訊號 pain:疼痛7步訊號 nopain:不痛7步訊號 find:畫圖title用,listCorrelation:計算重複點,listConvolution:計算重複點
#flag:在哪種特徵會出現的位置,list_findpeaks_point:記錄用特徵位置
def two_run_ncc(allname,pname,all,pain,nopain,find,listCorrelation,listConvolution,flag,list_findpeaks_point):

    #先記錄findpeak之前的位置
    listoddpoint=[]
    listoddpoint=listCorrelation[:]
    # if(listCorrelation==listoddpoint):
    #     print("可以")
    RX_pain_plot=NormaliseCrossCorrelation(all,pain)
    r_RX_pain_plot=NormaliseCrossConvolution(all,pain)
    #one_plot_start,one_plot_end=one_plot(len(all),len(pain))
    turn_plot_start,turn_plot_end,one_plot_start,one_plot_end,two_plot_start,two_plot_end=forward_record_plot(len(all),len(pain))
    #r_turn_plot_start,r_turn_plot_end,r_one_plot_start,r_one_plot_end,r_two_plot_start,r_two_plot_end=reverse_record_plot(len(all),len(pain))    


    #print(oneplot)
    #print(len(RX_pain_plot))
    #RX_pain_plot=NormaliseCrossCorrelation(listRX1,RX_pain)
    # for i in range(len(RX_pain_plot)):
    #     if(RX_pain_plot[i]<0):
    #         RX_pain_plot[i]=RX_pain_plot[i]*-1   

    #print(RX_pain_plot)
    


    # print("\n疼痛最大")
    # print(max(RX_pain_plot))
    # print("疼痛平均")
    # print(np.average(RX_pain_plot))

    
    #全部
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.6, 6))
    plt.subplots_adjust(hspace=0.3)

    # 第一個subplot顯示第一個圖表
    ax1.set_title(pname+"("+str(find)+"步"+")疼痛Pattern 對上"+allname+"的相關性")
    #ax1.set_xlim([5750, 7000])
    #ax1.plot(RX_nopain_plot)
    RX_pain_plot=np.array(RX_pain_plot)
    RX_pain_plot_peaks, _= find_peaks(RX_pain_plot,height=0.5,distance=80)

    ax1.plot(RX_pain_plot)
    ax1.plot(RX_pain_plot_peaks, RX_pain_plot[RX_pain_plot_peaks],"*")

    for i in range(len (one_plot_end)):
        if(i==0):
            ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
            ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
        else:
            ax1.axvline(x=one_plot_start[i],c='orange', linestyle="--")
            ax1.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    for i in range(len (two_plot_end)):
        if(i==0):
            ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
            ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")            
        else:
            ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
            ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")
    
    ax1.legend(loc='upper left', shadow=True) 

    # 第二個subplot顯示第二個圖表
    ax2.set_title(pname+"("+str(find)+"步"+")"+"Convolution對上"+allname)
    #ax2.set_xlim([5750, 7000])
    

    r_RX_pain_plot=np.array(r_RX_pain_plot)
    r_RX_pain_plot_peaks, _= find_peaks(r_RX_pain_plot,height=0.5,distance=80)

    ax2.plot(r_RX_pain_plot)
    ax2.plot(r_RX_pain_plot_peaks, r_RX_pain_plot[r_RX_pain_plot_peaks],"*")



    for i in range(len (one_plot_end)):
        if(i==0):
            ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
            ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
        else:
            ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
            ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    for i in range(len (two_plot_end)):
        if(i==0):
            ax2.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
            ax2.axvline(x=two_plot_end[i],c='r', linestyle="--")            
        else:
            ax2.axvline(x=two_plot_start[i],c='r', linestyle="--")
            ax2.axvline(x=two_plot_end[i],c='r', linestyle="--")

    ax2.legend(loc='upper left', shadow=True) 
    # 儲存並顯示圖表

    # folder_name = "example_folder"

    # if not os.path.exists(folder_name):
    #     os.mkdir(folder_name)

    # if pname == '右腳X特徵':
    #     if  allname=="右腳X":
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/右腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)

    #     if  allname == '左腳X':
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/左腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)

    # if pname == '左腳X特徵':
    #     if  allname=="右腳X":
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/右腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)

    #     if  allname == '左腳X':
    #         save_path = 'C:/Users/al574/Desktop/李O貴2/一級疼痛/右腳特徵/左腳'
    #         file_name = f"{pname}({find}步)疼痛Pattern{allname}"
    #         full_path = os.path.join(save_path, file_name)
    #         plt.savefig(full_path, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #         print("儲存成功:", full_path)
    # 儲存並顯示圖表
    folder_path = "C:\\Users\\al574\\Desktop\\輩分\\林O名\\0423test\\術前\\左右腳二級疼痛步態\\"+str(allname)
    # 创建指定路径下的文件夹（如果它不存在的话）
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)



    # 保存图像到指定的文件夹中
    filename = os.path.join(folder_path, "("+str(find)+"步"+")二級Pattern"+allname)
    #print(f"图片已保存为{filename}")
    plt.savefig(filename, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #plt.savefig("("+str(find)+"步"+")不痛Pattern"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    plt.show()




    
#==============================================================================================
    # #縮放
    # fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 6))
    # plt.subplots_adjust(hspace=0.3)


    # # 第一個subplot顯示第一個圖表
    # ax1.set_title("("+str(find)+"步"+")疼痛Pattern 對上"+allname+"的相關性")
    # ax1.set_xlim([5750, 7000])
    # #ax1.plot(RX_nopain_plot)
    # RX_pain_plot=np.array(RX_pain_plot)
    # RX_pain_plot_peaks, _= find_peaks(RX_pain_plot,height=0.5,distance=80)

    # ax1.plot(RX_pain_plot[:9600])
    # ax1.plot(RX_pain_plot_peaks, RX_pain_plot[RX_pain_plot_peaks],"*")
    # for i in range(len (one_plot_end)):
    #     if(i==0):
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
    #     else:
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    # for i in range(len (two_plot_end)):
    #     if(i==0):
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")            
    #     else:
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")
    # ax1.legend(loc='upper right', shadow=True) 
    
  

    # # 第二個subplot顯示第二個圖表
    # ax2.set_title("("+str(find)+"步"+")"+"Convolution對上"+allname)
    # ax2.set_xlim([5750, 7000])
    

    # r_RX_pain_plot=np.array(r_RX_pain_plot)
    # r_RX_pain_plot_peaks, _= find_peaks(r_RX_pain_plot,height=0.5,distance=80)

    # ax2.plot(r_RX_pain_plot[:9600])
    # ax2.plot(r_RX_pain_plot_peaks, r_RX_pain_plot[r_RX_pain_plot_peaks],"*")

    # for i in range(len (one_plot_end)):
    #     if(i==0):
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--",label='one')
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")            
    #     else:
    #         ax2.axvline(x=one_plot_start[i],c='orange', linestyle="--")
    #         ax2.axvline(x=one_plot_end[i],c='orange', linestyle="--")

    # for i in range(len (two_plot_end)):
    #     if(i==0):
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--",label='two')
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")            
    #     else:
    #         ax1.axvline(x=two_plot_start[i],c='r', linestyle="--")
    #         ax1.axvline(x=two_plot_end[i],c='r', linestyle="--")
            
    # ax2.legend(loc='upper right', shadow=True) 
    # # 儲存並顯示圖表
    # #plt.savefig("("+str(find)+"步"+")簡化"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    # plt.show()
#==============================================================================================

    #計算重複點
    for i in range(len(RX_pain_plot_peaks)):
        listCorrelation[RX_pain_plot_peaks[i]]=listCorrelation[RX_pain_plot_peaks[i]]+1


    for i in range(len(r_RX_pain_plot_peaks)):
        listConvolution[r_RX_pain_plot_peaks[i]]=listConvolution[r_RX_pain_plot_peaks[i]]+1 
    

    for i in range(len(listCorrelation)):
        if(listCorrelation[i]!=listoddpoint[i]):
            # print(listCorrelation[i])
            # print(listoddpoint[i])
            list_findpeaks_point[i]=flag
            # print(i)
            # print("flag")
            # print(flag)

    # for i in range(len(RX_pain_plot_peaks)):
    #     if(listCorrelation[i]>1) :
    #         print(listCorrelation[i])
    #         print(listoddpoint[i]) 



    




    #==================================不痛===========================================
    # if(find==1): 
    #     RX_nopain_plot=NormaliseCrossCorrelation(all,nopain)
    #     #RX_nopain_plot=NormaliseCrossCorrelation(listRX1,RX_nopain)
    #     for i in range(len(RX_nopain_plot)):
    #         if(RX_nopain_plot[i]<0):
    #             RX_nopain_plot[i]=RX_nopain_plot[i]*-1
    #     # print("\n不痛最大")
    #     # print(max(RX_nopain_plot))
    #     # print("不痛平均")
    #     # print(np.average(RX_nopain_plot)) 
    #     # print("=====================================================================")




    #     print(pname)
    #     plt.title("不痛Pattern 對上"+allname+"的相關性")
    #     plt.plot(RX_nopain_plot)
    #     #plt.axvline(x=flag,c='r')
    #     #plt.plot(Rc_turn_pic)
    #    # plt.plot(Rc_one_pic)
        
    #     #plt.savefig(pname+'不疼痛.png', dpi=750,facecolor='white')
    #     #plt.axvline(x=listflag_one[0],c='r')
    #     #plt.axvline(x=63,c='r')
    #     plt.show()
```

```python
R_flag=[]
#NCC
def NormaliseCrossCorrelation(a,v):
    M = len(a)
    N = len(v)

    if type(a) is list: 
        a = np.array(a)
    if type(v) is list: 
        v = np.array(v)
    
    v_mu = np.mean(v)

    ncc_series = []
    for i in range(M-N+1):
        _a = a[i:i+N]
        a_sum = _a - np.mean(_a)
        v_sum = v - v_mu
        
        nr = 1/N * np.sum(a_sum*v_sum) / np.sqrt(np.var(_a)*np.var(v))
        ncc_series.append(nr)

    return ncc_series


def NormaliseCrossConvolution(a,v):
    M = len(a)
    N = len(v)

    if type(a) is list: 
        a = np.array(a)
    if type(v) is list: 
        v = np.array(v)
    
    v_mean = np.mean(v)
    a_normalized = a - np.mean(a)
    v_normalized = v - v_mean
    #因為老師說要不一樣所以 mode=full 實際上是一樣的 因為是v翻轉不是實際波型所以相關度下降是正常的
    corr = np.convolve(a_normalized, v_normalized[::-1], mode='void')
    corr /= np.sqrt(np.sum(np.square(a_normalized)) * np.sum(np.square(v_normalized)))
    return corr



#A>B pname:圖片名稱備註 All:完整訊號 pain:疼痛7步訊號 nopain:不痛7步訊號 find:畫圖title用,listCorrelation:計算重複點,listConvolution:計算重複點
#flag:在哪種特徵會出現的位置,list_findpeaks_point:記錄用特徵位置
def after_one(allname,pname,all,pain,nopain,find,listCorrelation,listConvolution,flag,list_findpeaks_point):
    # #先記錄findpeak之前的位置
    # listoddpoint=[]
    # listoddpoint=listCorrelation[:]
    # if(listCorrelation==listoddpoint):
    #     print("可以")

    RX_pain_plot=NormaliseCrossCorrelation(all,pain)
    r_RX_pain_plot=NormaliseCrossConvolution(all,pain)

    
    #全部
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.6, 6))
    plt.subplots_adjust(hspace=0.3)

    # 第一個subplot顯示第一個圖表
    ax1.set_title(pname+"("+str(find)+"步"+")疼痛Pattern 對上"+allname+"的相關性")

    RX_pain_plot=np.array(RX_pain_plot)
    RX_pain_plot_peaks, _= find_peaks(RX_pain_plot,height=0.7,distance=80)

    ax1.plot(RX_pain_plot)
    ax1.plot(RX_pain_plot_peaks, RX_pain_plot[RX_pain_plot_peaks],"*")

    #ax1.legend(loc='upper left', shadow=True) 

    # 第二個subplot顯示第二個圖表
    ax2.set_title(pname+"("+str(find)+"步"+")"+"Convolution對上"+allname)
    #ax2.set_xlim([5750, 7000])
    

    r_RX_pain_plot=np.array(r_RX_pain_plot)
    r_RX_pain_plot_peaks, _= find_peaks(r_RX_pain_plot,height=0.7,distance=80)

    ax2.plot(r_RX_pain_plot)
    ax2.plot(r_RX_pain_plot_peaks, r_RX_pain_plot[r_RX_pain_plot_peaks],"*")
    #ax2.legend(loc='upper left', shadow=True)

    #儲存並顯示圖表
    folder_path = "C:\\Users\\al574\\Desktop\\輩分\\林O名\\0423test\\術後\\左右腳一級疼痛步態\\"+str(allname)
    # 创建指定路径下的文件夹（如果它不存在的话）
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)



    # 保存图像到指定的文件夹中
    filename = os.path.join(folder_path, "("+str(find)+"步"+")一級Pattern"+allname)
    #print(f"图片已保存为{filename}")
    plt.savefig(filename, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #plt.savefig("("+str(find)+"步"+")不痛Pattern"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    plt.show()
 
 
    

   # 計算重複點
    #print(RX_pain_plot_peaks)
    for i in range(len(RX_pain_plot_peaks)):
        listCorrelation[RX_pain_plot_peaks[i]]=listCorrelation[RX_pain_plot_peaks[i]]+1
        list_findpeaks_point[RX_pain_plot_peaks[i]]=flag

    for i in range(len(r_RX_pain_plot_peaks)):
        listConvolution[r_RX_pain_plot_peaks[i]]=listConvolution[r_RX_pain_plot_peaks[i]]+1


#A>B pname:圖片名稱備註 All:完整訊號 pain:疼痛7步訊號 nopain:不痛7步訊號 find:畫圖title用,listCorrelation:計算重複點,listConvolution:計算重複點
#flag:在哪種特徵會出現的位置,list_findpeaks_point:記錄用特徵位置
def after_two(allname,pname,all,pain,nopain,find,listCorrelation,listConvolution,flag,list_findpeaks_point):
    # #先記錄findpeak之前的位置
    # listoddpoint=[]
    # listoddpoint=listCorrelation[:]
    # if(listCorrelation==listoddpoint):
    #     print("可以")

    RX_pain_plot=NormaliseCrossCorrelation(all,pain)
    r_RX_pain_plot=NormaliseCrossConvolution(all,pain)

    
    #全部
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7.6, 6))
    plt.subplots_adjust(hspace=0.3)

    # 第一個subplot顯示第一個圖表
    ax1.set_title(pname+"("+str(find)+"步"+")疼痛Pattern 對上"+allname+"的相關性")

    RX_pain_plot=np.array(RX_pain_plot)
    RX_pain_plot_peaks, _= find_peaks(RX_pain_plot,height=0.7,distance=80)

    ax1.plot(RX_pain_plot)
    ax1.plot(RX_pain_plot_peaks, RX_pain_plot[RX_pain_plot_peaks],"*")

    #ax1.legend(loc='upper left', shadow=True) 

    # 第二個subplot顯示第二個圖表
    ax2.set_title(pname+"("+str(find)+"步"+")"+"Convolution對上"+allname)
    #ax2.set_xlim([5750, 7000])
    

    r_RX_pain_plot=np.array(r_RX_pain_plot)
    r_RX_pain_plot_peaks, _= find_peaks(r_RX_pain_plot,height=0.7,distance=80)

    ax2.plot(r_RX_pain_plot)
    ax2.plot(r_RX_pain_plot_peaks, r_RX_pain_plot[r_RX_pain_plot_peaks],"*")
    #ax2.legend(loc='upper left', shadow=True)

    #儲存並顯示圖表
    folder_path = "C:\\Users\\al574\\Desktop\\輩分\\林O名\\0423test\\術後\\左右腳二級疼痛步態\\"+str(allname)
    # 创建指定路径下的文件夹（如果它不存在的话）
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)



    # 保存图像到指定的文件夹中
    filename = os.path.join(folder_path, "("+str(find)+"步"+")二級Pattern"+allname)
    #print(f"图片已保存为{filename}")
    plt.savefig(filename, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    #plt.savefig("("+str(find)+"步"+")不痛Pattern"+allname, facecolor='white', bbox_inches='tight', pad_inches=0.1)
    plt.show()
 
 
    

   # 計算重複點
    #print(RX_pain_plot_peaks)
    for i in range(len(RX_pain_plot_peaks)):
        listCorrelation[RX_pain_plot_peaks[i]]=listCorrelation[RX_pain_plot_peaks[i]]+1
        list_findpeaks_point[RX_pain_plot_peaks[i]]=flag

    for i in range(len(r_RX_pain_plot_peaks)):
        listConvolution[r_RX_pain_plot_peaks[i]]=listConvolution[r_RX_pain_plot_peaks[i]]+1


```

```python
import statistics
#函式
#正規化
def Normalization(n):
    return (n-min(n))/(max(n)-min(n))

#函式0
#單一步態參數的標準差
def std(row,n):
    #n為取多少
    #m為n-1
    flag=0
    data=[]      #數據無正規化的標準差
    nordata=[]   #數據有正規化的標準差
    datanor=[]   #數據無正規化的標準差+正規化
    nordatanor=[]#數據有正規化的標準差+正規化
    m=n-1   
    for i in range(len(row) -m):
        b =np.array(row[i:n+ i])
        #print('無正規化數據:', b)
      
        st_dev = np.std(b)
        # print('數據無正規化標準差:',st_dev)
        # print('********************************************************')

        nb=Normalization(b)
        #print('正規化數據:', nb)
       
        nst_dev = np.std(nb)
        #print('正規化標準差:',nst_dev)
        #print('********************************************************')



        data.append(st_dev )
        nordata.append(nst_dev )
        
        
    datanor=Normalization(data)
    nordatanor=Normalization(nordata)
#[flag:判斷][data:無正規化的標準差][nordata:正規化數據後的標準差]
#[datanor:無正規化的標準差輸出做正規化][nordatanor:正規化數據後的標準差做正規化]
    return   flag,data,nordata,datanor,nordatanor

#函式1
#單一步態參數的變異度
def var(row,n):
    #n為取多少
    #m為n-1
    flag=1
    data=[]      #數據無正規化的標準差
    nordata=[]   #數據有正規化的標準差
    datanor=[]   #數據無正規化的標準差+正規化
    nordatanor=[]#數據有正規化的標準差+正規化
    m=n-1   
    for i in range(len(row) -m):
        b =np.array(row[i:n+ i])
        #print('無正規化數據:', b)
      
        d_var = np.var(b)
        # print('數據無正規化標準差:',st_dev)
        # print('********************************************************')

        nb=Normalization(b)
        #print('正規化數據:', nb)
       
        nd_var = np.var(nb)
        #print('正規化標準差:',nst_dev)
        #print('********************************************************')



        data.append(d_var )
        nordata.append(nd_var )
        
        
    datanor=Normalization(data)
    nordatanor=Normalization(nordata)
    #[flag:判斷][data:無正規化的標準差][nordata:正規化數據後的標準差]
    #[datanor:無正規化的標準差輸出做正規化][nordatanor:正規化數據後的標準差做正規化]
    return  flag,data,nordata,datanor,nordatanor


#函式2
#單一步態參數的變異度
def average(row,n):
    #n為取多少
    #m為n-1
    flag=2
    data=[]      #數據無正規化的標準差
    nordata=[]   #數據有正規化的標準差
    datanor=[]   #數據無正規化的標準差+正規化
    nordatanor=[]#數據有正規化的標準差+正規化
    m=n-1   
    for i in range(len(row) -m):
        b =np.array(row[i:n+ i])
        #print('無正規化數據:', b)
      
        d_var = np.average(b)
        # print('數據無正規化標準差:',st_dev)
        # print('********************************************************')

        nb=Normalization(b)
        #print('正規化數據:', nb)
       
        nd_var = np.average(nb)
        #print('正規化標準差:',nst_dev)
        #print('********************************************************')



        data.append(d_var )
        nordata.append(nd_var )
        
        
    datanor=Normalization(data)
    nordatanor=Normalization(nordata)
    #[flag:判斷][data:無正規化的標準差][nordata:正規化數據後的標準差]
    #[datanor:無正規化的標準差輸出做正規化][nordatanor:正規化數據後的標準差做正規化]
    return  flag,data,nordata,datanor,nordatanor


#函式3
#單一步態參數的平均
def mean(row,n):
    #n為取多少
    #m為n-1
    flag=3
    data=[]      #數據無正規化的標準差
    nordata=[]   #數據有正規化的標準差
    datanor=[]   #數據無正規化的標準差+正規化
    nordatanor=[]#數據有正規化的標準差+正規化
    m=n-1   
    for i in range(len(row) -m):
        b =np.array(row[i:n+ i])
        #print('無正規化數據:', b)
      
        d_mean = np.mean(b)
        # print('數據無正規化標準差:',st_dev)
        # print('********************************************************')

        nb=Normalization(b)
        #print('正規化數據:', nb)
       
        nd_mean = np.mean(nb)
        #print('正規化標準差:',nst_dev)
        #print('********************************************************')



        data.append(d_mean )
        nordata.append(nd_mean )
        
        
    datanor=Normalization(data)
    nordatanor=Normalization(nordata)
    #[flag:判斷][data:無正規化的標準差][nordata:正規化數據後的標準差]
    #[datanor:無正規化的標準差輸出做正規化][nordatanor:正規化數據後的標準差做正規化]
    return  flag,data,nordata,datanor,nordatanor

#轉彎函式
def data_p_turn(choose,flag,dan,d,nd,dn,ndn):
     #=========================================================================================
                #原始程式碼
                # plt.title(dan+"數據無正規化的標準差")
                # plt.plot(d)
                # plt.show()

                # plt.title(dan+"數據[有]正規化的標準差")
                # plt.plot(nd)
                # plt.show()
    #=========================================================================================
   if (flag==0):
        if(choose==0):
            
            plt.title(dan+"(轉彎)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_turn)
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(轉彎)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_turn)
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show()
        #=========================================================================================     
                    #原始程式碼
                    # plt.title(dan+"數據[有]正規化的標準差+輸出正規化畫圖")
                    # plt.plot(ndn)
                    # #plt.savefig( dan + '2.png', dpi=750,facecolor='white')
                    # plt.show() 
        #=========================================================================================
   if (flag==1):
        if(choose==0):
            
            plt.title(dan+"(轉彎)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_turn)
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(轉彎)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_turn)
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 

            #=========================================================================================
                    #原始程式碼
                    # plt.title(dan+"變異度的輸出正規化畫圖")
                    # plt.plot(d) 
                    # #plt.savefig(dan + '變異度的輸出正規化畫圖.png', dpi=750,facecolor='white')
                    # plt.show() 
                    
                    # plt.title(dan+"數據[有]正規化的變異度+輸出正規化畫圖")
                    # plt.plot(ndn)
                    # #plt.savefig( dan + '2.png', dpi=750,facecolor='white')
                    # plt.show() 
            #=========================================================================================

   if (flag==2):
            if(choose==0):
                
                plt.title(dan+"(轉彎)平均average的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Rc_turn)
                #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
            
            if(choose==1):
                
                plt.title(dan+"(轉彎)平均average的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Lc_turn)
                #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
   if (flag==3):
            if(choose==0):
                
                plt.title(dan+"(轉彎)均值mean的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Rc_turn)
                #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
            
            if(choose==1):
                
                plt.title(dan+"(轉彎)均值mean的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Lc_turn)
                #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 

#一級疼痛函式
def data_p_one(choose,flag,dan,d,nd,dn,ndn):
     #=========================================================================================
                #原始程式碼
                # plt.title(dan+"數據無正規化的標準差")
                # plt.plot(d)
                # plt.show()

                # plt.title(dan+"數據[有]正規化的標準差")
                # plt.plot(nd)
                # plt.show()
    #=========================================================================================
   if (flag==0):
        if(choose==0):
            
            plt.title(dan+"(疼痛)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_one)
            plt.plot(Rc_two,color='r')
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(疼痛)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_one)
            plt.plot(Lc_two,color='r')
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show()
        #=========================================================================================     
                    #原始程式碼
                    # plt.title(dan+"數據[有]正規化的標準差+輸出正規化畫圖")
                    # plt.plot(ndn)
                    # #plt.savefig( dan + '2.png', dpi=750,facecolor='white')
                    # plt.show() 
        #=========================================================================================
   if (flag==1):
        if(choose==0):
            
            plt.title(dan+"(疼痛)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_one)
            plt.plot(Rc_two,color='r')
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(疼痛)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_one)
            plt.plot(Lc_two,color='r')
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 

            #=========================================================================================
                    #原始程式碼
                    # plt.title(dan+"變異度的輸出正規化畫圖")
                    # plt.plot(d) 
                    # #plt.savefig(dan + '變異度的輸出正規化畫圖.png', dpi=750,facecolor='white')
                    # plt.show() 
                    
                    # plt.title(dan+"數據[有]正規化的變異度+輸出正規化畫圖")
                    # plt.plot(ndn)
                    # #plt.savefig( dan + '2.png', dpi=750,facecolor='white')
                    # plt.show() 
            #=========================================================================================

   if (flag==2):
            if(choose==0):
                
                plt.title(dan+"(疼痛)平均average的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Rc_one)
                plt.plot(Rc_two,color='r')
                #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
            
            if(choose==1):
                
                plt.title(dan+"(疼痛)平均average的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Lc_one)
                plt.plot(Lc_two,color='r')
                #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
   if (flag==3):
            if(choose==0):
                
                plt.title(dan+"(疼痛)均值mean的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Rc_one)
                plt.plot(Rc_two,color='r')
                #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
            
            if(choose==1):
                
                plt.title(dan+"(疼痛)均值mean的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Lc_one)
                plt.plot(Lc_two,color='r')
                #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 

#二級疼痛函式
def data_p_two(choose,flag,dan,d,nd,dn,ndn):
    
   if (flag==0):
        if(choose==0):
            
            plt.title(dan+"(二級疼痛)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_two)
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(二級疼痛)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_two)
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show()
       
   if (flag==1):
        if(choose==0):
            
            plt.title(dan+"(二級疼痛)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_two)
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(二級疼痛)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_two)
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 


   if (flag==2):
            if(choose==0):
                
                plt.title(dan+"(二級疼痛)平均average的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Rc_two)
                #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
            
            if(choose==1):
                
                plt.title(dan+"(二級疼痛)平均average的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Lc_two)
                #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
   if (flag==3):
            if(choose==0):
                
                plt.title(dan+"(二級疼痛)均值mean的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Rc_two)
                #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 
            
            if(choose==1):
                
                plt.title(dan+"(二級疼痛)均值mean的輸出正規化畫圖")
                plt.plot(dn)
                plt.plot(Lc_two)
                #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
                plt.show() 

#三級疼痛函式
def data_p_three(choose,flag,dan,d,nd,dn,ndn):
    
   if (flag==0):
        if(choose==0):
            
            plt.title(dan+"(三級疼痛)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_three)
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(三級疼痛)標準差std的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_three)
            #plt.savefig(dan + '標準差std的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show()
       
   if (flag==1):
        if(choose==0):
            
            plt.title(dan+"(三級疼痛)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_three)
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(三級疼痛)變異度var的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_three)
            #plt.savefig(dan + '變異度var的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 


   if (flag==2):
        if(choose==0):
            
            plt.title(dan+"(三級疼痛)平均average的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_three)
            #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(三級疼痛)平均average的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_three)
            #plt.savefig(dan + '平均average的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
   if (flag==3):
        if(choose==0):
            
            plt.title(dan+"(三級疼痛)均值mean的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Rc_three)
            #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 
        
        if(choose==1):
            
            plt.title(dan+"(三級疼痛)均值mean的輸出正規化畫圖")
            plt.plot(dn)
            plt.plot(Lc_three)
            #plt.savefig(dan + '均值mean的輸出正規化畫圖.png', dpi=750,facecolor='white')
            plt.show() 


```
